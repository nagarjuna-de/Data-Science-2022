```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('img/m.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
```

```python
# Mask out other than blue
lower_bounds = (110,50,200)
upper_bounds = (160,255,255)
mask = cv2.inRange(hsv_img, lower_bounds, upper_bounds)
cv2.imshow('only blue', mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
