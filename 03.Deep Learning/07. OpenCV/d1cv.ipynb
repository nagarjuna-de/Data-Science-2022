### Image Annotations

```python
img = cv2.imread('img/day1.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
copy_img = rgb_img.copy()
cv2.rectangle(copy_img, (500,500), (600,600), (0,255,0), cv2.FILLED)
plt.figure(figsize=(20,16))
plt.imshow(copy_img)
```

```python
copy_img = rgb_img.copy()
cv2.putText(copy_img, 'This is my text', (200,500), cv2.FONT_HERSHEY_PLAIN, 3, (0,255,0),3)
plt.figure(figsize=(20,16))
plt.imshow(copy_img)
```

```python
copy_img = rgb_img.copy()
plt.figure(figsize=(20,15))

cv2.circle(copy_img, (200,200),80, (0,127,127), cv2.FILLED)
plt.imshow(copy_img)
```

```python
plt.figure(figsize=(20,15))

cv2.circle(copy_img, (200,200),90, (255,0,0), 2)
plt.imshow(copy_img)
```

### HSV and Masks

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from cv2 import imshow
bgr_flowers = cv2.imread('img/purple-flowers.jpg')
rgb_flowers = cv2.cvtColor(bgr_flowers, cv2.COLOR_BGR2RGB)
hsv_flowers = cv2.cvtColor(bgr_flowers, cv2.COLOR_BGR2HSV)
imshow('test', bgr_flowers)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

lower_bounds = (110,50,200)
upper_bounds = (160,255,255)
mask = cv2.inRange(hsv_flowers, lower_bounds, upper_bounds)
cv2.imshow('masked gray', mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(mask, cmap='gray')
```

```python
masked_flowers = bgr_flowers.copy()
masked_flowers[mask==0] = [0,0,0]
cv2.imshow('masked out', masked_flowers)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
