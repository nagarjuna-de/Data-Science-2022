```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def imshow(img):
    show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(15,16))
    plt.imshow(show_img)
    plt.xticks(color='White')
    plt.yticks(color='White')
```

```python
coins = cv2.imread('img/coins5.jpg')
imshow(coins)
```

```python
grayscale_coins = cv2.cvtColor(coins,cv2.COLOR_BGR2GRAY)

retval, threshold = cv2.threshold(grayscale_coins, 200,255, cv2.THRESH_BINARY_INV)

imshow(threshold)
```

```python
countours, heirarchy = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```
