```python
url ="https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data"
import pandas as pd
pd.read_csv(url, header=None)
```

```python
columns = ["animal name",
"hair",
"feathers",
"eggs",
"milk",
"airborne",
"aquatic",
"predator",
"toothed",
"backbone",
"breathes",
"venomous",
"fins",
"legs",
"tail",
"domestic",
"catsize",
"type"]
```

```python
df= pd.read_csv(url, header=None, names=columns)
df
```

```python
df = df.drop(["animal name"], axis=1) 
df
```

```python
X = df.drop(["type"], axis=1)
y = df.type
```

```python
X
```

```python
import matplotlib.pyplot as plt
for col in df.columns:
    plt.figure()
    plt.hist(df[col])
    plt.title(col, color = "White")
    plt.xticks(color = "White")
    plt.yticks(color = "White")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf = dt_clf.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15,15))
plot_tree(dt_clf, 
        filled=True, 
        rounded=True,
        class_names=["Mamal","Bird","Reptile","Fish","Amphibian","Bug","Invertebrate"],
        feature_names=X.columns)
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(dt_clf,X_test,y_test,display_labels=["Mamal","Bird","Reptile","Fish","Amphibian","Bug","Invertebrate"])
plt.xticks(color="White")
plt.yticks(color="White")
```
